```python
import pandas as pd
import random
import re
from collections import Counter

data = pd.read_csv(r"C:\Users\vynzo\Downloads\artworks.csv", dtype=object, encoding='latin-1')

new = data['Title'].str.split(',', expand = True)

title = new[0].str.split(' ', expand = True)

res = list(title[0]) + list(title[1]) + list(title[2]) + list(title[3]) + list(title[4]) + list(title[5]) + list(title[6]) + list(title[7]) + list(title[8]) + list(title[9]) + list(title[10])

words = list(filter(None, res))

words
```

```python
first_word = re.sub(r'[^a-zA-Z]', "", random.choice(words))
second_word = re.sub(r'[^a-zA-Z]', "", random.choice(words))
third_word = re.sub(r'[^a-zA-Z]', "", random.choice(words))
fourth_word = re.sub(r'[^a-zA-Z]', "", random.choice(words))
name = first_word + " " + second_word + " " + third_word + " " + fourth_word
print(name)
```
